```
#Jong Ha Lee, jlee616@jhu.edu, jlee616
print('Jong Ha Lee, jlee616@jhu.edu, jlee616')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.linear_model as lm

! if [ ! -e orings.csv ]; \
then wget https://raw.githubusercontent.com/bcaffo/ds4bme_intro/master/data/orings.csv; \
fi;

df = pd.read_csv("orings.csv")

tot = df['Total']
#make tot binary
for i in range(len(tot)):
  #define o-ring failure (total>0) as 1
  if tot[i] > 0:
    tot[i]=1

sns.scatterplot(df['Temperature'],tot)

temp = df[['Temperature']]

lr = lm.LogisticRegression(fit_intercept=True, C = 100000)
fit = lr.fit(temp, tot)

beta0, beta1 = [fit.intercept_[0], fit.coef_[0][0]]
[beta0, beta1]

n = 1000
xplot = np.linspace(50, 85, n)
eta = beta0 + beta1 * xplot
p = 1 / (1 + np.exp(-eta))

sns.lineplot(xplot,p)

t=74
eta2 = beta0 + beta1 * t
prob = 1 / (1 + np.exp(-eta2))
print('Probability of failure for temperature of 74 degrees: ',prob)
```
